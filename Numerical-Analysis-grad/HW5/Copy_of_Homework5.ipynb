```python
# CP.4.5.8

data = np.array([
    [1, 8.0],
    [2, 12.3],
    [3, 15.5],
    [4, 16.8],
    [5, 17.1],
    [6, 15.8],
    [7, 15.2],
    [8, 14.0]
])

t = data[:,0]
y = data[:,1]

f = lambda c: np.array([
    c[0] * t[0] * np.exp(c[1] * t[0]) - y[0],
    c[0] * t[1] * np.exp(c[1] * t[1]) - y[1],
    c[0] * t[2] * np.exp(c[1] * t[2]) - y[2],
    c[0] * t[3] * np.exp(c[1] * t[3]) - y[3],
    c[0] * t[4] * np.exp(c[1] * t[4]) - y[4],
    c[0] * t[5] * np.exp(c[1] * t[5]) - y[5],
    c[0] * t[6] * np.exp(c[1] * t[6]) - y[6],
    c[0] * t[7] * np.exp(c[1] * t[7]) - y[7]
])


J = lambda c :[
    [t[0] * np.exp(c[1] * t[0]), c[0] * t[0]**2 * np.exp(c[1] * t[0])],
    [t[1] * np.exp(c[1] * t[1]), c[0] * t[1]**2 * np.exp(c[1] * t[1])],
    [t[2] * np.exp(c[1] * t[2]), c[0] * t[2]**2 * np.exp(c[1] * t[2])],
    [t[3] * np.exp(c[1] * t[3]), c[0] * t[3]**2 * np.exp(c[1] * t[3])],
    [t[4] * np.exp(c[1] * t[4]), c[0] * t[4]**2 * np.exp(c[1] * t[4])],
    [t[5] * np.exp(c[1] * t[5]), c[0] * t[5]**2 * np.exp(c[1] * t[5])],
    [t[6] * np.exp(c[1] * t[6]), c[0] * t[6]**2 * np.exp(c[1] * t[6])],
    [t[7] * np.exp(c[1] * t[7]), c[0] * t[7]**2 * np.exp(c[1] * t[7])]
]


c = np.array( [ 10., 0. ])
lbda = 1

fc = f( c )

for _ in range(0,10):
  Jc = J(c)
  JcTJc_plI = np.transpose(Jc) @ Jc + lbda * np.eye(np.shape(Jc)[1])
  c = c - np.linalg.solve( JcTJc_plI, np.transpose(Jc) @ fc )
  fc = f(c)
  nrm_r = np.linalg.norm( fc, 2 )
  print( "x = [ ", f"{c[0]:.10f}", ",", f"{c[1]:.10f}", " ]", "    nrm_r = ", f"{nrm_r:.16f}")

print("RMSE = ", f"{np.linalg.norm( fc, 2 ) / sqrt( t.shape[0] ) :.4f}")

xxx = np.linspace(0.9, 8.1, 100)
yyy = c[0] * xxx * np.exp(c[1] * xxx)

plt.plot(xxx, yyy, '--b', label=f'{c[0]:7.5f} * t * e ^ {c[1]:7.5f} * t')
plt.plot(t, y, 'ro')
[plt.text(t[i], y[i], f"{data[i,0]:.0f}") for i in range(data.shape[0])]
plt.legend()
plt.xlabel('hour (hr)')
plt.ylabel('concentration (ng/ml)')
plt.xlim([0, 10])
plt.ylim([0, 20])
plt.grid()
plt.show()

# from google.colab import files
# plt.savefig('exercise_cp_4_5_08_________sauer3____plot.pdf');
# files.download('exercise_cp_4_5_08_________sauer3____plot.pdf')
```
